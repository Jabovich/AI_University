```python
import numpy as np
from matplotlib import pyplot as plt

def print_img(img: np.array):
    plt.imshow(img, interpolation='nearest')

def print_bl_white_img(img):
    plt.imshow(img, cmap='Greys')

# 1.Найти индексы ненулевых элементов в заданном массиве.
a1 = np.nonzero([0, 2, 0, 7, 0, 8, 6, 44, 0, 4])
print(a1, "\n")

# 2.Создать NxN единичную матрицу.
a2 = np.eye(6)
print(a2, "\n")

# 3.Создать массив nxnxn со случайными значениями
a3 = np.random.randint(1, 100, (5, 5))
print(a3, "\n")

# 4.Создать вектор (одномерный массив) размера n, заполненный нулями
n4 = 5
a4 = np.zeros(n4)
print(a4, "\n")

# 5.Создать вектор размера 1n, заполненный нулями, но пятый элемент равен 1
n5 = 5
a5 = np.zeros(n5)
a5[4] = 1
print(a5, "\n")

# 6.Создать вектор со значениями от n до m
n6 = 1
m6 = 10
a6 = np.random.randint(n6, m6, 5)
print(a6, "\n")

# 7.Развернуть одномерный массив (первый становится последним)
a7 = np.arange(10)
a7 = a7[::-1]
print(a7, "\n")

# 8.Создать матрицу с 0 внутри, и 1 на границах
a8 = np.zeros(10)
a8[0] = 1
a8[9] = 1
print(a8, "\n")

# 9.Создать nxn матрицу и заполнить её в шахматном порядке
n9 = 8
a9 = np.zeros((n9, n9), dtype=int)
a9[1::2, ::2] = 1
a9[::2, 1::2] = 1
print(a9, "\n")

# 10.Заменить максимальный элемент на ноль
a10 = np.random.randint(1, 100, 5)
print(a10)
a10[a10.argmax()] = 0
print(a10, "\n")

# 11.Найти ближайшее к заданному значению число в заданном массиве
n11 = 50
a11 = np.random.randint(1, 100, 5)
print(a11)
index = (np.abs(a11-n11)).argmin()
print(a11[index], "\n")

# 12.Дан четырехмерный массив, посчитать сумму по последним двум осям
a12 = np.random.randint(0, 10, (4, 4, 4, 4))
print(a12)
sum = a12.reshape(a12.shape[:-2] + (-1,)).sum(axis=-1)
print(sum, "\n")

# 13.Определить, есть ли в двумерном массиве нулевые столбцы
a13 = np.random.randint(0, 2, (2, 5))
print(a13)
zero_column = (~a13.any(axis=0)).any()
print(zero_column, "\n")

# 14.Отнять среднее из каждой строки в матрице
a14 = np.random.randint(1, 10, (5, 5))
print(a14)
middle = a14 - a14.mean(axis=1, keepdims=True)
print(middle, "\n")

# 15.Создайте вектор с элементами от 10 до 50
a15 = np.arange(10, 50)
print(a15, "\n")

# 16.Создайте матрицу (5, 5), заполненую от 0 до 6
a16 = np.random.randint(0, 6, (5, 5))
print(a16, "\n")

# 17.Умножьте матрицу размерности (5, 3) на (3, 2)
a17 = np.dot(np.random.randint(1, 10, (5, 3)), np.random.randint(1, 10, (3, 2)))
print(a17, "\n")

# 18.Создайте матрицу (10, 10) так, чтобы на границе были 0, а внтури 1
a18 = np.zeros((10, 10))
a18[1:-1,1:-1] = 1
print(a18, "\n")

# 19.Для заданного числа найдите ближайший к нему элемент в векторе
n19 = 10
a19 = np.random.randint(1, 21, 5)
print(a19)
index = (np.abs(a19-n19)).argmin()
print(a11[index], "\n")

# 20.Дан трёхмерный массив "Цветное изображение", содержащий изображение, размера (height, width, numChannels), а также вектор коэффициентов длины numChannels.
# Сложить каналы изображения с указанными весами, и вернуть результат в виде матрицы размера (height, width).
# Реальное изображение не считываем, заполняем трехмерный массив случайными значениями.
# Преобразуйте цветное изображение в оттенки серого, использовав коэффициенты np.array([0.299, 0.587, 0.114]).
height = 20
width = 25
numChannels = 3
img = np.random.randint(0, 255, height*width*numChannels).reshape(height, width, numChannels)
print_img(img)
numChannels = np.array([0.299, 0.587, 0.114])
bw = (img.dot(numChannels))
print_bl_white_img(bw)
```
