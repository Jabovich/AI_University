# 2. Линейная диаграмма

4. Постройте несколько вертикальных линий разных типов, ширины и цветов на одном графике
5. Измените непрозрачность строки в Matplotlib
6. Увеличьте толщину линии с помощью Matplotlib
7. Как заполнить области между несколькими строками в Matplotlib?

```python
import numpy as np
import matplotlib.pyplot as plt

#1.Постройте горизонтальную линию в Matplotlib (использовать plt.axhline)
plt.axhline(y = 0.5, color = 'green', linestyle = '-')

#2.Постройте несколько горизонтальных линий разных типов, ширины и цветов на одном графике
plt.axhline(y = 1, color = 'blue', linestyle = '--')
plt.axhline(y = 1.5, color = 'red', linestyle = ':')
plt.axhline(y = 2, color = 'black', linestyle = '-', linewidth=5)
```

```python
import numpy as np
import matplotlib.pyplot as plt

#3.Постройте вертикальную линию в Matplotlib
plt.axvline(x = 0.5, color = 'red', linestyle = '--')

#4.Постройте несколько вертикальных линий разных типов, ширины и цветов на одном графике
plt.axvline(x = 1, color = 'purple', linestyle = '--', linewidth=5)
plt.axvline(x = 1.5, color = 'blue', linestyle = '-')
plt.axvline(x = 2, color = 'green', linestyle = ':', linewidth=3)

#5.Измените непрозрачность строки в Matplotlib
plt.axvline(x = 2.5, color = 'black', linestyle = '-', linewidth=5, alpha=0.4)

#6.Увеличьте толщину линии с помощью Matplotlib
plt.axvline(x = 3, color = 'yellow', linestyle = '-', linewidth=10)

#7.Как заполнить области между несколькими строками в Matplotlib?
#plt.fill_between(x, y, color='') для заполнения между двумя горизонтальными линиями
#plt.fill_betweenx(x, y, color='') для заполнения между двумя вертикальными линиями
```
